Dataset: Digits dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
import ipdb
# Config the matlotlib backend as plotting inline in IPython
%matplotlib inline
```

Cargamos el dataset

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
n_clases = len(np.unique(y))
print("Numero de muestras: ",len(y))
print("Numero de clases: ",n_clases)
print("Dimensiones de X:",X.shape)
```

Visualizacion en 2 dimensiones
-------------------------------

```python
def plot_clusters(X,K,k_ids,centroids=[]):
    pca = PCA(n_components=2)
    X2d = pca.fit_transform(X)
    if centroids!=[]:
        cent2d = pca.transform(centroids)

    plt.figure(figsize=(10,10))
    colors = 'rgmybck'
    
    for i in range(K):
        if sum(k_ids==i)==0:
            continue
        plt.scatter(X2d[k_ids==i,0],X2d[k_ids==i,1],c=colors[i%len(colors)],marker='o',alpha=0.8)
        plt.hold(True)
    if centroids!=[]:
        for i in range(K):
            plt.plot(cent2d[i,0],cent2d[i,1],'b+',linewidth=10)
    plt.title("Visualizacion para %i clusters en Digits Dataset" % K)
    plt.show()
```

```python
plot_clusters(X,n_clases,y)
```

Implementación de algoritmo KMeans
-----------------------------------

```python
def KMeans(X,K,nro_iteraciones=100,eps=1e-4):
    N,M = X.shape
    k_ids = np.zeros(N)
    
    # Inicializar centroides
    centroids = np.random.random([K,M])
    
    for _iter in range(nro_iteraciones):
        # Asignar cluster id del centroide mas cercano
        for i in range(N):
            xi = X[i,:]
            dist = np.zeros(K) # distance xi->mu_
            for k in range(K):
                mu_k = centroids[k,:]
                # xi - mu_k : np.linalg.norm(xi-mu_k)
                dist[k] = np.linalg.norm(xi-mu_k)
            k_ids[i] = dist.argmin()
        
        # Actualizar Centroides
        for k in range(K):
            xk = X[k_ids==k,:] # todas las muestras del cluster k
            new_muk = xk.mean(axis=0) # nuevo centroide de cluster k
            if sum(np.isnan(new_muk))==0:
                centroids[k,:] = new_muk
            else:
                centroids[k,:] = np.random.random(M)

        if _iter%10 == 0:
            print("Iteracion %i..." % _iter)
    return k_ids,centroids

        
```

Probar algoritmo

```python
K=5
k_ids,centroids = KMeans(X,K,10)

plot_clusters(X,K,k_ids,centroids)
```
