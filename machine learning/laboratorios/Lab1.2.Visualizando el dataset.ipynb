Laboratorio 1. Parte 1. Visualización de dataset de Digitos Manuscritos
------------------------

El objetivo de esta parte es cargar y visualizar el dataset de [dígitos manuscritos](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html) disponible en Scikit-Learn. Esta es una pequeña porción del conocido dataset [MNIST](http://yann.lecun.com/exdb/mnist/).
Cada imagen es de 8x8 pixeles, con 1797 muestras distribuidas entre todas las clases (digitos del 0 al 9).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Config the matlotlib backend as plotting inline in IPython
%matplotlib inline
```

Cargamos el dataset

```python
digits = datasets.load_digits()

X, y = digits.data, digits.target

print(X.shape)
print(y.shape)
```

Visualizamos las primeras 9 imagenes de un dígito

```python
digit = 8 # cambiar para visualizar otro digito

Xdig = X[y==digit,:]

plt.figure(figsize=(9,9))
for i in range(1,10):
    plt.subplot(3,3,i)
    plt.axis('off')
    plt.imshow(Xdig[i-1,:].reshape((8,8)),cmap='gray')
plt.show()  
    
```
